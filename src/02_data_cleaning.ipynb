```python
import pandas as pd
import zipfile

zip_path = "../data/raw/Accidents(16-23).zip"

with zipfile.ZipFile(zip_path) as z:
    with z.open('Accidents(16-23).csv') as f:
        df = pd.read_csv(f, nrows=100000)
```

```python
#check missing percentages
missing_percent = df.isnull().mean()*100

missing_percent.sort_values(ascending=False).head(20)
```

```python
# Drop columns having >40% missing values.

drop_cols = missing_percent[
    missing_percent > 40
].index

df.drop(columns=drop_cols,
        inplace=True)

print(drop_cols)
```

```python
#Fill categorical columns.

cat_cols = df.select_dtypes(
    include='object'
).columns

for col in cat_cols:
    df[col].fillna(
        df[col].mode()[0],
        inplace=True
    )
```

```python
#Remove duplicates.

df.drop_duplicates(inplace=True)

df.shape
```

```python
for col in ['Temperature(F)',
            'Humidity(%)',
            'Visibility(mi)',
            'Wind_Speed(mph)']:

    df[col] = df[col].fillna(df[col].median())
```

```python
print(df[['Temperature(F)',
          'Humidity(%)',
          'Visibility(mi)',
          'Wind_Speed(mph)']].isnull().sum())
```

```python
#Save cleaned data

df.to_csv(
    "../data/processed/cleaned_data.csv",
    index=False
)
```
