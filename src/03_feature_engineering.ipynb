```python
import pandas as pd

df = pd.read_csv(
    "../data/processed/cleaned_data.csv"
)
```

```python
print(df[['Temperature(F)',
          'Humidity(%)',
          'Visibility(mi)',
          'Wind_Speed(mph)']].isnull().sum())
```

```python
#conversion of date columns to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'], format='ISO8601')

df['End_Time'] = pd.to_datetime(df['End_Time'], format='ISO8601')
```

```python
#Hour
df['hour'] = df['Start_Time'].dt.hour
```

```python
#Day of week

df['day_of_week'] = (
    df['Start_Time']
    .dt.dayofweek
)
```

```python
#Month

df['month'] = (
    df['Start_Time']
    .dt.month
)
```

```python
#Rush hour

df['is_rush_hour'] = (
    df['hour']
    .apply(
        lambda x: 1 if (
            7 <= x <= 9 or
            17 <= x <= 19
        )
        else 0
    )
)
```

```python
#Night indicator

df['is_night'] = (
    df['Sunrise_Sunset']
    .map({
        'Night':1,
        'Day':0
    })
)
```

```python
#Duration of accident in minutes

df['duration'] = (
    pd.to_datetime(df['End_Time'])
    -
    pd.to_datetime(df['Start_Time'])
).dt.total_seconds()/60
```

```python
#Save

df.to_csv(
    "../data/processed/featured_data.csv",
    index=False
)
```
